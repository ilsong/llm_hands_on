```python
import torch
import torch.nn.functional as F

# 예시 데이터
logits = torch.randn(2, 3, 50257) # (Batch, Seq_Len, Vocab_Size)
targets = torch.randint(0, 50257, (2, 3)) # (Batch, Seq_Len)

def calculate_loss(logits, targets):
    # [문제 2-1] CrossEntropyLoss 계산을 위해 logits과 targets를 평탄화(Flatten)하세요.
    # (힌트: (2, 3, 50257) -> (6, 50257), (2, 3) -> (6)으로 변환)
    logits_flat = torch.flatten(logits, start_dim=0) # 여기에 코드를 작성하세요
    targets_flat = torch.flatten(targets, start_dim=2) # 여기에 코드를 작성하세요

    # [문제 2-2] F.cross_entropy를 사용하여 최종 손실값을 계산하세요.
    loss = F.cross_entropy(logits_flat, targets_flat) # 여기에 코드를 작성하세요
    return loss

print(f"계산된 손실값: {calculate_loss(logits, targets)}")
```
